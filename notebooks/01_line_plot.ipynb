# ðŸ§© Notebook-01: Line Plot with Matplotlib

```python
# Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import sys
from pathlib import Path

# âœ… Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import line_plot, multi_line_plot

print("ðŸ“ˆ Line Plots using Matplotlib\n")
```

```python
# Load Dataset
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
df.head()
```

```python
# Single Line Plot â€“ Total Units Sold Over Time
df_grouped = df.groupby("Month")["Units Sold"].sum().reset_index()

line_plot(
    x=df_grouped["Month"],
    y=df_grouped["Units Sold"],
    title="ðŸ“ˆ Total Monthly Units Sold (All Products)",
    xlabel="Month",
    ylabel="Units Sold",
    label="All Products",
    color="blue",
    marker="o"
)
```

```python
# Multi-Line Plot â€“ Product-Wise Sales Trend

df_pivot = df.pivot(index="Month", columns="Product", values="Units Sold").reset_index()
df_pivot.head()

multi_line_plot(
    df=df_pivot,
    x_col="Month",
    y_cols=["Laptop", "Tablet", "Smartphone", "Monitor"],
    labels=["Laptop", "Tablet", "Smartphone", "Monitor"],
    colors=["blue", "green", "red", "orange"],
    markers=["o", "s", "^", "D"],
    title="ðŸ“Š Product-Wise Monthly Units Sold",
    xlabel="Month",
    ylabel="Units Sold"
)
```

```python
# Line Styles & Customization
line_plot(
    x=df_grouped["Month"],
    y=df_grouped["Units Sold"],
    title="Dashed Line with Triangle Markers",
    xlabel="Month",
    ylabel="Units Sold",
    label="Total Sales",
    color="purple",
    linestyle="--",
    marker="^"
)
```
