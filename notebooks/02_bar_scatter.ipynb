# üß© Notebook-02: Bar and Scatter Plots with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import bar_plot, grouped_bar_plot, scatter_plot

print("üìä Bar and Scatter Plots using Matplotlib\n")
```

```python
# üóÇÔ∏è 1. Load Dataset
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
df.head()
```

```python
# üì¶ 2. Vertical Bar Plot ‚Äì Total Units Sold per Product
product_totals = df.groupby("Product")["Units Sold"].sum().reset_index()

bar_plot(
    categories=product_totals["Product"],
    values=product_totals["Units Sold"],
    title="üì¶ Total Units Sold per Product",
    xlabel="Product",
    ylabel="Units Sold"
)
```

```python
# üìä 3. Horizontal Bar Plot ‚Äì Revenue per Product
revenue_totals = df.groupby("Product")["Revenue"].sum().reset_index()

bar_plot(
    categories=revenue_totals["Product"],
    values=revenue_totals["Revenue"],
    horizontal=True,
    title="üí∞ Total Revenue per Product",
    xlabel="Revenue ($)",
    ylabel="Product",
    color="salmon"
)
```

```python
# ü™Ñ 4. Grouped Bar Plot ‚Äì Product-Wise Sales Over Months
df_grouped = df.groupby(["Month", "Product"])["Units Sold"].sum().unstack().reset_index()
data_dict = {
    product: df_grouped[product].tolist() for product in df_grouped.columns[1:]
}

grouped_bar_plot(
    x_labels=df_grouped["Month"].tolist(),
    data_dict=data_dict,
    title="üìä Monthly Units Sold by Product",
    xlabel="Month",
    ylabel="Units Sold"
)
```

```python
# üéØ 5. Scatter Plot ‚Äì Units Sold vs Revenue
scatter_plot(
    x=df["Units Sold"],
    y=df["Revenue"],
    title="üîç Revenue vs. Units Sold",
    xlabel="Units Sold",
    ylabel="Revenue"
)
```

```python
# üé® 6. Advanced Scatter ‚Äì Color + Size Mapping
df["ColorMap"] = df["Units Sold"]
df["BubbleSize"] = np.sqrt(df["Revenue"])

scatter_plot(
    x=df["Units Sold"],
    y=df["Revenue"],
    color_values=df["ColorMap"],
    size=df["BubbleSize"],
    cmap="viridis",
    title="üß† Revenue vs. Units Sold (Colored & Scaled)",
    xlabel="Units Sold",
    ylabel="Revenue",
    use_colorbar=True,
    edgecolor="black"
)
```

- ‚úÖ Built vertical, horizontal, and grouped bars
