# üß© Notebook-03: Histograms and Pie Charts with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import histogram_plot, pie_chart

print("üìä Histograms and Pie Charts using Matplotlib\n")
```

```python
# üóÇÔ∏è 1. Load Dataset
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
df.head()
```

```python
# üìà 2. Histogram ‚Äì Distribution of Units Sold
histogram_plot(
    data=df["Units Sold"],
    bins=10,
    color="skyblue",
    edgecolor="black",
    title="üìà Distribution of Units Sold",
    xlabel="Units Sold",
    ylabel="Frequency"
)
```

```python
# üìä 3. Histogram ‚Äì Revenue Distribution with Density
histogram_plot(
    data=df["Revenue"],
    bins=15,
    color="orange",
    edgecolor="white",
    alpha=0.8,
    density=True,
    title="üí∞ Revenue Distribution (Density)",
    xlabel="Revenue"
)
```

```python
# ü•ß 4. Pie Chart ‚Äì Product Sales Proportion
product_totals = df.groupby("Product")["Units Sold"].sum().reset_index()

pie_chart(
    labels=product_totals["Product"],
    values=product_totals["Units Sold"],
    colors=["#66b3ff", "#99ff99", "#ffcc99", "#ff9999"],
    explode=[0.05]*4,
    title="ü•ß Units Sold Share by Product"
)
```

```python
# üí∞ 5. Pie Chart ‚Äì Revenue Proportion by Product
revenue_totals = df.groupby("Product")["Revenue"].sum().reset_index()

pie_chart(
    labels=revenue_totals["Product"],
    values=revenue_totals["Revenue"],
    colors=["#8ecae6", "#219ebc", "#ffb703", "#fb8500"],
    explode=[0.05]*4,
    title="üí∞ Revenue Share by Product"
)
```

## ‚úÖ Summary
- ‚úÖ Used `histogram_plot()` for frequency and density plots
- ‚úÖ Visualized revenue and unit distribution with bins
- ‚úÖ Created styled pie charts for proportion analysis
- ‚úÖ Introduced explode, custom colors, and shadows
