# üß© Notebook-05: Customization and Styling with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import line_plot
from theme_utils import apply_dark_theme, apply_corporate_theme, apply_minimal_theme, reset_theme

print("üé® Customization and Styling in Matplotlib\n")
```

```python
# üóÇÔ∏è 1. Load Dataset

df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
monthly = df.groupby("Month")["Units Sold"].sum().reset_index()
```

```python
# üé® 2. Default Theme (Baseline)

reset_theme()

line_plot(
    x=monthly["Month"],
    y=monthly["Units Sold"],
    title="üß™ Default Theme: Units Sold",
    xlabel="Month",
    ylabel="Units Sold",
    label="Units Sold",
    color="mediumblue",
    marker="o"
)
```

```python
# üñ§ 3. Dark Theme
apply_dark_theme()

line_plot(
    x=monthly["Month"],
    y=monthly["Units Sold"],
    title="üåô Dark Theme: Units Sold",
    xlabel="Month",
    ylabel="Units Sold",
    label="Units Sold",
    color="cyan",
    marker="D"
)
```

```python
# üßë‚Äçüíº 4. Corporate Theme
apply_corporate_theme()

line_plot(
    x=monthly["Month"],
    y=monthly["Units Sold"],
    title="üè¢ Corporate Theme: Units Sold",
    xlabel="Month",
    ylabel="Units Sold",
    label="Units Sold",
    color="#0077b6",
    marker="s"
)
```

```python
# ‚ú® 5. Minimal Theme
apply_minimal_theme()

line_plot(
    x=monthly["Month"],
    y=monthly["Units Sold"],
    title="üßº Minimal Theme: Units Sold",
    xlabel="Month",
    ylabel="Units Sold",
    label="Units Sold",
    color="indigo",
    marker="^"
)
```

## ‚úÖ Summary
- ‚úÖ Applied 4 custom themes using rcParams: dark, corporate, minimal, and reset
- ‚úÖ Encapsulated themes inside `theme_utils.py`
- ‚úÖ Custom styling improves readability, aesthetics, and presentation
