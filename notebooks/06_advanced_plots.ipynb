```python
# üß© Notebook-06: Advanced Plots with Matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import fill_between_plot, log_scale_plot, twin_axes_fill_plot

print("üî¨ Advanced Plots using Matplotlib\n")
```

```python
# üóÇÔ∏è 1. Load Dataset
df_sales = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
df_stocks = pd.read_csv(PROJECT_ROOT / "datasets" / "stock_prices.csv")
```

```python
# üåà 2. Fill Between ‚Äì Units Sold Over Time
monthly_sales = df_sales.groupby("Month")["Units Sold"].sum().reset_index()

fill_between_plot(
    x=monthly_sales["Month"],
    y1=monthly_sales["Units Sold"],
    title="üìâ Area Under Total Units Sold",
    xlabel="Month",
    ylabel="Units Sold",
    color="lightcoral",
    label="Units Sold"
)
```

```python
# üî¨ 3. Log Scale ‚Äì Revenue Distribution (Log Y-axis)
revenue_by_order = df_sales["Revenue"].sort_values().reset_index(drop=True)

log_scale_plot(
    x=range(len(revenue_by_order)),
    y=revenue_by_order,
    log_axis="y",
    title="üìà Revenue Distribution (Log-Scaled)",
    xlabel="Transactions",
    ylabel="Revenue ($)",
    color="darkorange",
    label="Revenue"
)
```

```python
# üìä 4. Log-Log Plot ‚Äì Stock High vs Volume
df_tesla = df_stocks[df_stocks["Stock"] == "TSLA"]

log_scale_plot(
    x=df_tesla["Volume"],
    y=df_tesla["High"],
    log_axis="both",
    title="‚öôÔ∏è TSLA: High Price vs Volume (Log-Log)",
    xlabel="Volume",
    ylabel="High Price",
    color="green",
    marker="o",
    label="TSLA"
)
```

```python
# ü™Ñ 5. Twin Axis Fill ‚Äì Units Sold and Revenue
monthly_agg = df_sales.groupby("Month")[["Units Sold", "Revenue"]].sum().reset_index()

twin_axes_fill_plot(
    x=monthly_agg["Month"],
    y1=monthly_agg["Units Sold"],
    y2=monthly_agg["Revenue"],
    label1="Units Sold",
    label2="Revenue",
    xlabel="Month",
    ylabel1="Units Sold",
    ylabel2="Revenue ($)",
    title="ü™Ñ Area Plot: Units Sold vs Revenue",
    color1="teal",
    color2="darkviolet"
)
```

# ‚úÖ Summary
- ‚úÖ Used `fill_between_plot()` to visualize trends over time
- ‚úÖ Applied log scale using `log_scale_plot()` for better distribution views
- ‚úÖ Created a log-log plot for stock data
- ‚úÖ Used `twin_axes_fill_plot()` to overlay two scaled trends visually
