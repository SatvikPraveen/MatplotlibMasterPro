# üß© Notebook-07: Annotations and Highlights with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import annotate_point, highlight_region, label_line

print("üñçÔ∏è Annotations and Highlights in Matplotlib\n")
```

```python
# üóÇÔ∏è 1. Load Dataset
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
monthly = df.groupby("Month")[["Units Sold", "Revenue"]].sum().reset_index()
```

```python
# üìç 2. Annotate a Specific Point ‚Äì Highest Units Sold
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(monthly["Month"], monthly["Units Sold"], marker='o', color='teal')

# Annotate max point
max_idx = monthly["Units Sold"].idxmax()
x_max = monthly["Month"][max_idx]
y_max = monthly["Units Sold"][max_idx]

annotate_point(ax, x=x_max, y=y_max, text="üî∫ Highest Sales", textcolor='teal')
ax.set_title("üìç Annotate Peak Units Sold")
ax.set_xlabel("Month")
ax.set_ylabel("Units Sold")
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
# üå´Ô∏è 3. Highlight a Region ‚Äì Holiday Spike (e.g., Nov‚ÄìDec)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(monthly["Month"], monthly["Units Sold"], color='darkorange', marker='s')
highlight_region(ax, x_start="2023-11", x_end="2023-12", color="gold", alpha=0.4, label="üéÅ Holiday Spike")
ax.set_title("üå´Ô∏è Highlight Holiday Sales Region")
ax.set_xlabel("Month")
ax.set_ylabel("Units Sold")
ax.legend()
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
# üè∑Ô∏è 4. Label Line Directly ‚Äì Revenue vs Units Sold (with twin axes)
fig, ax1 = plt.subplots(figsize=(10, 5))
x = monthly["Month"]
y1 = monthly["Units Sold"]
y2 = monthly["Revenue"]

ax1.plot(x, y1, color="blue", marker="o")
label_line(ax1, x=list(x), y=list(y1), text="Units Sold", location="right", color="blue", offset=(0, 5))
ax1.set_ylabel("Units Sold", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.plot(x, y2, color="green", marker="^")
label_line(ax2, x=list(x), y=list(y2), text="Revenue", location="right", color="green", offset=(0, -10))
ax2.set_ylabel("Revenue", color="green")
ax2.tick_params(axis='y', labelcolor="green")

ax1.set_xlabel("Month")
ax1.set_title("üè∑Ô∏è Label Lines Directly (with Twin Axes)")
ax1.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

plt.show()
```

# ‚úÖ Summary
- ‚úÖ Used `annotate_point()` to highlight max points
- ‚úÖ Applied `highlight_region()` to mark time intervals
- ‚úÖ Labeled curves directly with `label_line()` for clarity
