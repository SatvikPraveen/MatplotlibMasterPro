```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import imshow_matrix, plot_image, grid_heatmap

print("üñºÔ∏è Image and Grid Visualization in Matplotlib\n")
```

```python
# 1. Matrix Visualization ‚Äì Simulated Confusion Matrix
# Simulate a confusion matrix
conf_matrix = np.array([
    [42, 2, 3],
    [4, 35, 1],
    [2, 5, 41]
])
class_labels = ['Class A', 'Class B', 'Class C']

imshow_matrix(
    matrix=conf_matrix,
    cmap='Blues',
    title="üß™ Simulated Confusion Matrix",
    xlabel="Predicted",
    ylabel="Actual"
)
```

```python
# üî• 2. Grid Heatmap ‚Äì Annotated Weather Map
# Simulate city-level temperatures
data = np.random.uniform(15, 40, (4, 6)).round(1)
cities = ['Austin', 'Chicago', 'NYC', 'San Francisco']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

grid_heatmap(
    data=data,
    row_labels=cities,
    col_labels=months,
    cmap="Oranges",
    annot=True,
    fmt=".1f",
    title="üî• Simulated City Temperature Grid (¬∞C)"
)
```
