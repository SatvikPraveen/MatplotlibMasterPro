# üß© Notebook-10: Exporting and Styling Plots in Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import save_plot, set_global_style

print("üñºÔ∏è Exporting & Styling with Matplotlib\n")
```

```python
# üé® 1. Set Global Style
set_global_style(style_name="seaborn-v0_8-whitegrid", font="DejaVu Sans", size=12)
```

```python
# üìä 2. Load Data
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
monthly = df.groupby("Month")[["Units Sold", "Revenue"]].sum().reset_index()
```

```python
# üìà 3. Generate Plot with Style
fig, ax = plt.subplots(figsize=(10, 5))
x = monthly["Month"]
y = monthly["Revenue"]

ax.plot(x, y, marker="o", color="purple")
ax.set_title("üñºÔ∏è Monthly Revenue", fontsize=14)
ax.set_xlabel("Month")
ax.set_ylabel("Revenue ($)")
ax.tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

```python
# üíæ 4. Save Plot in Multiple Formats
save_plot(fig, filename="monthly_revenue", folder=PROJECT_ROOT / "exports")
```
