# üß© Notebook-11: Animated Line Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import animate_line_plot

print("üéûÔ∏è Animated Line Plots using Matplotlib Animation")
```

```python
# üìÇ 1. Load and Prepare Data
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
monthly = df.groupby("Month")[["Units Sold", "Revenue"]].sum().reset_index()

x = monthly["Month"]
y = monthly["Revenue"]
```

```python
# üìà 2. Create Animation ‚Äì Monthly Revenue Growth
animate_line_plot(
    x=np.arange(len(x)),
    y=y.values,
    xlabel="Month Index",
    ylabel="Revenue",
    title="üìà Revenue Growth Over Time (Animated)",
    interval=400,
    color="green"
)
```

```python
# üîÅ 3. Animate Dual Line Plot
from plot_utils import animate_dual_line_plot

animate_dual_line_plot(
    x=monthly["Month"],
    y1=monthly["Units Sold"],
    y2=monthly["Revenue"],
    xlabel="Month",
    ylabel="Value",
    labels=("Units Sold", "Revenue"),
    title="üîÅ Animate Units Sold and Revenue"
)
```

```python
# üìä 3. Animated Bar Chart
from plot_utils import animate_bar_chart

product = df.groupby("Product")["Revenue"].sum().sort_values()
animate_bar_chart(
    categories=product.index.tolist(),
    values=product.values.tolist(),
    xlabel="Product",
    ylabel="Revenue",
    title="üìä Animated Product-wise Revenue"
)
```

```python
# üîµ 4. Animated Scatter Growth
from plot_utils import animate_scatter_growth

animate_scatter_growth(
    x=df["Units Sold"],
    y=df["Revenue"],
    xlabel="Units Sold",
    ylabel="Revenue",
    title="üîµ Revenue vs Units Sold Growth"
)
```

```python
# üì¶ 5. Export Animations as .mp4

from plot_utils import (
    save_line_animation,
    save_dual_line_animation,
    save_bar_animation,
    save_scatter_animation
)

# üîÉ Export: Single Line Plot
save_line_animation(
    x=np.arange(len(x)),
    y=y.values,
    filename="../exports/revenue_growth.mp4"
)

# üîÅ Export: Dual Line Plot
save_dual_line_animation(
    x=np.arange(len(monthly["Month"])),
    y1=monthly["Units Sold"].values,
    y2=monthly["Revenue"].values,
    labels=("Units Sold", "Revenue"),
    filename="../exports/units_revenue_growth.mp4"
)

# üìä Export: Bar Chart Animation
save_bar_animation(
    categories=product.index.tolist(),
    values=product.values.tolist(),
    filename="../exports/product_revenue_bars.mp4"
)

# üîµ Export: Scatter Animation
save_scatter_animation(
    x=df["Units Sold"].values,
    y=df["Revenue"].values,
    filename="../exports/revenue_vs_units_scatter.mp4"
)
```
