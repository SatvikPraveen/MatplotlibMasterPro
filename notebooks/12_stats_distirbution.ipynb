# ðŸ§© Notebook-12: Stats & Distribution Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
from pathlib import Path

# âœ… Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

from plot_utils import (
    plot_histogram_with_stats,
    plot_boxplot,
    plot_violinplot
)

print("ðŸ“ˆ Statistical & Distribution Plots with Matplotlib\n")
```

```python
# ðŸ“‚ 1. Load and Prepare Data
df = pd.read_csv(PROJECT_ROOT / "datasets" / "sales_data.csv")
monthly = df.groupby("Month")[["Units Sold", "Revenue"]].sum().reset_index()

units = monthly["Units Sold"]
revenue = monthly["Revenue"]

# ðŸ“Š 2. Histogram with Mean, Median, Std
plot_histogram_with_stats(
    data=units,
    bins=10,
    title="ðŸ“Š Distribution of Units Sold",
    xlabel="Units Sold"
)
```

```python
# ðŸ“¦ 3. Boxplot of Units vs Revenue
plot_boxplot(
    data_dict={"Units Sold": units, "Revenue": revenue},
    title="ðŸ“¦ Boxplot: Units Sold vs Revenue",
    ylabel="Values"
)
```

```python
# ðŸŽ» 4. Violin Plot (Optional but Beautiful!)
plot_violinplot(
    data_dict={"Units Sold": units, "Revenue": revenue},
    title="ðŸŽ» Violin Plot: Units Sold vs Revenue",
    ylabel="Values"
)
```

```python
# ðŸ“ˆ 5. KDE-style Line Plot (Bonus!)
plt.figure(figsize=(8, 4))
units_sorted = np.sort(units)
density = np.exp(-0.5 * ((units_sorted - np.mean(units_sorted)) / np.std(units_sorted)) ** 2)
density = density / (np.std(units_sorted) * np.sqrt(2 * np.pi))
plt.plot(units_sorted, density, label="Estimated KDE", color="purple")
plt.title("ðŸ“ˆ KDE-style Plot for Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Density")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.legend()
plt.show()
```
